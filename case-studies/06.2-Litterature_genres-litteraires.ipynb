```python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

openai.api_key  = os.getenv('OPENAI_API_KEY')

client = openai.OpenAI()
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0
    )
    return response.choices[0].message.content
```

```python
prompt = f"""
Génère une liste des genres littéraires majeurs. 
Pour chaque genre, indique son nom, une brève définition, ses principales caractéristiques 
(par exemple : narration à la première personne, présence de dialogues, visée argumentative), 
et un ou deux exemples d'œuvres représentatives. 
Fournis les données au format JSON avec les clés suivantes : 
genre_id, nom, définition, caractéristiques, œuvres_exemplaires.
"""

response = get_completion(prompt)
print(response)
```

```python
import openai


def get_completion(prompt, model="gpt-4.1"):
    messages = [
        {
            "role": "system", 
            "content": "Tu es un assistant expert en littérature française."
        },
        {
            "role": "user", 
            "content": prompt
        }]
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
        max_tokens=1500
    )
    return response.choices[0].message.content


prompt = """
Présente les principaux genres littéraires français de manière structurée et pédagogique. Pour chaque genre, indique :

1. Une brève définition,
2. Les formes principales associées à ce genre,
3. Ses caractéristiques stylistiques ou thématiques,
4. Des exemples d’œuvres et d’auteurs emblématiques.

Structure la réponse de manière lisible avec des titres et des puces. 
Termine par une section sur les genres hybrides ou contemporains si possible.

Fournis les données au format JSON 
"""


response = get_completion(prompt)
print(response)
```
